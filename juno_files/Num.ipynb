```python
from scipy.stats import multivariate_normal as mvn
```

```python
def create_mult_norm_dist(means, cov_mat, n, feat_names):
    scores = mvn.rvs(mean=means, cov=cov_mat, size=n)
    df = pd.DataFrame(data=scores, columns=feat_names)
    return df
```

```python
cov = np.array([[1, 0.9,.7, .6],[.9,1.,.5,.5],[0.7,.5,1.,.5],[0.6,.5,.5,1]])
cov
```

```python
scores = mvn.rvs(mean = [60.,80.,60.,60.], cov=cov, size=1000)
```

```python
np.corrcoef(scores.T)
```

```python
df = pd.DataFrame(data = scores, columns = ["Math", "Science","History", "Art"])
df.head()
```
